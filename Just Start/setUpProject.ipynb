```python
import ollama

# Define the model name
model_name = "gemma3:1b"

# Function to generate restaurant name
def generate_restaurant_name(cuisine):
    prompt = f"Generate a unique restaurant name for a {cuisine} cuisine."
    response = ollama.chat(model=model_name, messages=[{"role": "user", "content": prompt}])
    return response["message"]["content"]

# Function to generate menu
def generate_menu(restaurant_name):
    prompt = f"Generate a menu with 5 dishes for the restaurant {restaurant_name}."
    response = ollama.chat(model=model_name, messages=[{"role": "user", "content": prompt}])
    return response["message"]["content"]

# Main function
def generate_restaurant_and_menu(cuisine):
    restaurant_name = generate_restaurant_name(cuisine)
    menu = generate_menu(restaurant_name)
    return {"restaurant_name": restaurant_name, "menu": menu}

# Example usage
cuisine = "Italian"
result = generate_restaurant_and_menu(cuisine)
print(result)
```

```python

# Import necessary modules
from langchain.tools import WikipediaQueryRun
from langchain.utilities import WikipediaAPIWrapper
from langchain.agents import initialize_agent, AgentType
from langchain.llms import Ollama
from langchain.tools import Tool

wikipedia = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())


llm = Ollama(model="gemma3:1b")  


agent = initialize_agent(
    tools=[wikipedia],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True, 
    handle_parsing_errors=True
)


query = "Elon musk age in 2025?"
response = agent.run(query)


print("\n📖 Wikipedia Answer:\n")
print(response)
```

🛠️ Key Components of Agents
LLM (Language Model) – Core reasoning engine (e.g., GPT-4, Gemma, Llama).

Tools (Functions/APIs) – External resources the agent can call (e.g., Calculator, Wikipedia, Web Scraper).

Memory – Stores past interactions for better contextual understanding.

Output Parser – Converts model responses into structured actions.

Prompt Template – Defines how the agent interacts and makes decisions.
